neste dataset tenho:

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter 

df = pd.read_csv("https://raw.githubusercontent.com/beatrizadm/data-analysis/master/data/top50-2019.csv")
df.head()
```

Quais os cantores mais recorrentes nas top songs?

```
bar_plot = dict(Counter(df['Artist.Name'].values).most_common(5))
plt.bar(*zip(*bar_plot.items()), color=(0.9,0.8,0.2,0.9))
plt.show()
```

O que faz as m√∫sicas de ed sheeran terem tanto sucesso em um s√≥ ano?

```
#ordenando por popularidade
df = df.sort_values('Popularity', ascending=False)
ed_sheeran = df.loc[df["Artist.Name"]=="Ed Sheeran"]
ed_sheeran.head()
```

Podemos perceber que suas m√∫sicas em geral s√£o mais ac√∫sticas do que "faladas".

```
# Data
track_name = ['Antisocial', 'Beautiful People', "I Don't Care", 'Cross Me']
data=pd.DataFrame({'x': track_name, 'Acousticness': ed_sheeran['Acousticness..'], 'Speechiness': ed_sheeran['Speechiness.'] })
 
# multiple line plot
plt.plot('x', 'Acousticness', data=data, marker='o', markerfacecolor='#561fff', markersize=12, color='#8f93ff', linewidth=4)
plt.plot('x', 'Speechiness', data=data, marker='o', markerfacecolor='purple', markersize=12, color='#e99dff', linewidth=4)
plt.legend()
```

Apenas na m√∫sica "Beautiful People" esse padr√£o n√£o ocorre. 
J√° √© conhecido que as m√∫sicas de Ed Sheeran tem bastante musicalidade e com ritmo, agora vamos ver a rela√ß√£o de beats por minuto, dan√ßabilidade e energia.

```
x = track_name
y = ed_sheeran['Beats.Per.Minute']
z = ed_sheeran['Danceability']
w = ed_sheeran['Valence.']

plt.scatter(x, y, s=y*3, c="green", alpha=0.7, label='Beats por minuto')
plt.scatter(x, z, s=z*3, c="blue", alpha=0.7, label='Dan√ßabilidade')
plt.scatter(x, w, s=w*3, c="red", alpha=0.7, label='Valencia')

plt.legend()
plt.show()
```

Logo percebemos que por mais que a m√∫sica tenha muitos beats por minuto e relativa dan√ßabilidade, n√£o significa que ela ter√° alta val√™ncia, tendo em vista que suas m√∫sicas mais reflexivas e que passam mensagens que nem sempre s√£o positivas.
